## Check Password Strength Using ML algorithms

```python
import pandas as pd
import numpy as np
import seaborn as sns
import random
sns.set()
```

```python
dataset = pd.read_csv('data.csv',',',error_bad_lines=False)
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
passwords=np.array(dataset)
```

```python

random.shuffle(passwords) #shuffling randomly for robustness
```

```python
y=[labels[1] for labels in passwords]
```

```python
y
```

```python
X=[i[0] for i in passwords]
```

```python
sns.set_style('dark')
sns.countplot(x='strength',data=dataset)
```

```python
X
```

```python
def password_s(inputs):
    char=[]
    for i in inputs:
        char.append(i)
    return char
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
model=TfidfVectorizer(tokenizer=password_s)
X=model.fit_transform(X)
```

```python
## Logistics Regression

from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
```

```python
regressor=LogisticRegression(penalty='l2',multi_class='ovr')
regressor.fit(X_train,y_train)
```

```python
print(regressor.score(X_test,y_test))
```

```python
New_X=np.array(["%@123abcd"])
New_X=model.transform(New_X)
prediction=regressor.predict(New_X)
print(prediction)
```
